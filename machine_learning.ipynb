```python
import pandas as pd

clients = pd.read_csv("data/clients.csv")
sinistres = pd.read_csv("data/sinistres.csv")
contacts = pd.read_csv("data/contacts.csv")

df = contacts.merge(sinistres, on=["sinistre_id", "client_id"], how="left")
df = df.merge(clients, on="client_id", how="left")
```

```python
print(df.shape)
```

```python
print(df.head())
```

```python
df["date_contact"] = pd.to_datetime(df["date_contact"])
df["date_ouverture"] = pd.to_datetime(df["date_ouverture"])
df["date_inscription"] = pd.to_datetime(df["date_inscription"])
```

```python
today = pd.Timestamp.today().normalize()

df["anciennete_client_jours"] = (today - df["date_inscription"]).dt.days
df["anciennete_sinistre_jours"] = (today - df["date_ouverture"]).dt.days
df["mois_contact"] = df["date_contact"].dt.month
df["jour_semaine_contact"] = df["date_contact"].dt.dayofweek
```
